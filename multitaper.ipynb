```python
import os 
import mne
from tqdm import tqdm
from pipe import EEGDataset
from pipe import ML_EEG
from sklearn.svm import SVC
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
mne.set_log_level('ERROR')

root = "/mnt/w/Dementia_Project/Datasets/ds004504/"
dataset = EEGDataset(root,  epoch_length=38.0, overlap=19.0)
```

# Alzheimer and Common

```python
XX_train_data = dataset.experiments.train.ad_cn.data
yy_train = dataset.experiments.train.ad_cn.label
dataset.experiments.train.ad_cn.get_distribution()

XX_train = []

for epoch in XX_train_data:
    XX_train.append(epoch.getmne_multitaper(index=0))
    
XX_test_data = dataset.experiments.test.ad_cn.data
yy_test = dataset.experiments.test.ad_cn.label
dataset.experiments.test.ad_cn.get_distribution()

XX_test = []

for epoch in XX_test_data:
    XX_test.append(epoch.getmne_multitaper(index=0))


lda = LinearDiscriminantAnalysis()
ML_EEG(lda, XX_train, yy_train, XX_test, yy_test, label="LDA")

print("################################################")

svc = SVC(kernel='poly')
ML_EEG(svc, XX_train, yy_train, XX_test, yy_test, label="Support Vector Machine")

print("################################################")
knn =  KNeighborsClassifier(n_neighbors=13)
ML_EEG(knn, XX_train, yy_train, XX_test, yy_test, label="KNeighborsClassifier")

print("################################################")

rf = RandomForestClassifier(n_estimators=3)  
ML_EEG(rf, XX_train, yy_train, XX_test, yy_test, label="RandomForestClassifier")
```
